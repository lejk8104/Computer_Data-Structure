```
class User:
  
  def __init__(self, name):
    self.name = name
    self.posts = [ ] # post는 리스트로 관리
    self.followees = [ ] # 팔로워도 리스트로 관리
    
  def get_name(self):
    return self.name
  
  def get_posts(self):
    return self.posts
  
  def get_followees(self):
    return self.followees
  
  def add_post(self, post):
    self.posts.append(post)
  
  def add_followee(self, followee):
    self.followees.append(followee)
      
  
class Post:
  
  def __init__(self, text, date):
    self.text = text
    self.date = date
    self.like = 0
    
  def get_text(self):
    return self.text
  
  def get_date(self):
    return self.date
  
  def get_like(self):
    return self.like
  
  def increase_like(self):
    self.like += 1

# first user
u_1 = User("Kim")

p_1 = Post("I like Python.", "20180305")
for i in range(0, 5):
  p_1.increase_like()
u_1.add_post(p_1)

p_2 = Post("I like Java as well.", "20180517")
for i in range(0, 2):
  p_2.increase_like()
u_1.add_post(p_2)


# second user
u_2 = User("Lee")

p_3 = Post("Python function", "20181023")
for i in range(0, 2):
  p_3.increase_like()
u_2.add_post(p_3)


# third user
u_3 = User("Park")

p_4 = Post("Python vs. Java", "20160907")
for i in range(0, 10):
  p_4.increase_like()
u_3.add_post(p_4)

p_5 = Post("Soccer", "20171013")
u_3.add_post(p_5)

p_6 = Post("Programming", "20181125")
for i in range(0, 4):
  p_6.increase_like()
u_3.add_post(p_6)

# forth user
u_4 = User("Choi")

p_7 = Post("Python and AI", "20150811")
for i in range(0, 100):
  p_7.increase_like()
u_4.add_post(p_7)

p_8 = Post("Vacation", "20181128")
for i in range(0, 300):
  p_8.increase_like()
u_4.add_post(p_8)

# set followees
u_1.add_followee(u_2)
u_1.add_followee(u_3)
u_1.add_followee(u_4)


# quiz 1.
print(u_1.get_name())
print("------------------------------------")

# quiz 2. u_1의 팔로워하는 인원의 수를 출력하시오
followees = u_1.get_followees()        #U1의 팔로워 정의
print(len(followees))                  #팔로워의 숫자 출력
print("------------------------------------")

# quiz 3.
for followee in followees:             
  print(followee.get_name())
print("------------------------------------")

# quiz 4.                              #u1의팔로워중 -part이라는 이름이 들어간 분들의 포스트의 수를 출력하시오
for followee in followees:             #u1의 팔로워 숫자만큼 반복
  if followee.get_name() == "Park":    #팔로워의 이름이 park일 경우
    posts = followee.get_posts()       #그포스트를 정의
    print(len(posts))                  #포스트의 숫자 출력
print("------------------------------------")

# quiz 5.                              #u1의 팔로워 중 park이라는 사람의 좋아요의 개수를 출력하시오
for followee in followees:             
  if followee.get_name() == "Park":
    posts = followee.get_posts()
    like_count = 0                    #좋아요 초기화 0
    for post in posts:
      like_count += post.get_like()   #하나씩 증감 like
    print(like_count)
print("------------------------------------")

# quiz 6.                              #u1의 팔로워의 포스터 중 python이 들어가면 출력하시오
for followee in followees:
  posts = followee.get_posts()
  for post in posts:
    text = post.get_text()
    if text.find("Python") >= 0:
      print(text)
print("------------------------------------")

# quiz 7.                               #python이 들어간 사람의 이름과 내용을 출력하시오
for followee in followees:
  posts = followee.get_posts()
  name = followee.get_name()
  for post in posts:
    text = post.get_text()
    if text.find("Python") >= 0:
      print("[" + name + "] " + text)
print("------------------------------------")

# quiz 8.
for followee in followees:
  posts = followee.get_posts()
  for post in posts:
    date = post.get_date()
    year = date[:4]
    year = int(year)
    if year >= 2017 and year <= 2018:
      text = post.get_text()
      like = post.get_like()
      like = str(like)
      print(text + " [" + like + "]")
print("------------------------------------")

# quiz 9.                       #?
all_posts = []
for followee in followees:
  posts = followee.get_posts()
  for post in posts:
    all_posts.append(post)

size = len(all_posts) 
for i in range(0, size-1):
  for j in range(0, size-1-i):
    if all_posts[j].get_date() < all_posts[j+1].get_date():
      temp = all_posts[j]
      all_posts[j] = all_posts[j+1]
      all_posts[j+1] = temp
for post in all_posts:
  print(post.get_text())
print("------------------------------------")
  
# quiz 9-1.                      #?
all_posts = []
for followee in followees:
  posts = followee.get_posts()
  for post in posts:
    all_posts.append(post)

limit = 3
size = len(all_posts) 
for i in range(0, limit):
  for j in range(0, size-1-i):
    if all_posts[j].get_date() > all_posts[j+1].get_date():
      temp = all_posts[j]
      all_posts[j] = all_posts[j+1]
      all_posts[j+1] = temp
      
for i in range(len(all_posts)-1, len(all_posts)-1-limit, -1):
  print(all_posts[i].get_text())
```
