```

# 교수님 강의자료 : 모범답안

class Node:                                              # 재귀함수 안써. Queue와 stack이 재귀함수의 기능을 대신하여 안써도 됨
  def __init__(self, data):
    self.data = data
    self.neighbors = []
    self.visited = False                                 #Flag
    
  def add_neighbor(self, neighbor):
    self.neighbors.append(neighbor)

  def set_visited(self, visited):                        #Flag를 올리거나 내리거나 설정하는 것
    self.visited = visited

  def get_data(self):
    return self.data           

  def get_neighbors(self):                               #이웃 내놔봐 이때 리스트로 나옴
    return self.neighbors

  def get_visited(self):
    return self.visited                                  #방문할 수 있을까?


class Graph:
  def __init__(self):                                    #이웃들은 노드가 갖고있기 때문에 (트리랑 같다? 자식들을 부모가 갖고있어서)
    self.nodes =  

  def add_node(self, node):
    self.nodes.append(node)
    
  def bfs(self):
    if len(self.nodes) > 0:                              # 반복작업을 위한 노드세팅 초기화.(모두 깃발내려)
      for node in self.nodes:
        node.set_visited(False)
        
    queue = []
    self.nodes[0].set_visited(True)                      #첫번째노드(A)를 지정. (A노드 깃발올려)
    print(self.nodes[0].get_data(), end="")              #노드A 출력
    queue.append(self.nodes[0])                          #queue list에 node A 추가 (queue의 기능1 각 노드들을 기억한다.)
    while len(queue) > 0:                                #queue list가 0이 되기전까지 행동반복.(queue의 기늗2 너비우선탐색의 End condition 조정)
      current = queue.pop(0)                             #처음에 들어온 데이터 = current 이젠 nodes[1]이 기준점(dequeue)
      neighbors = current.get_neighbors()
      for neighbor in neighbors:
        if not neighbor.get_visited():                   #이웃들에게도 같은작업 반복. 너 방문안했어? 이제부터 Ture(깃발올려)
          neighbor.set_visited(True)
          print(" - " + neighbor.get_data(), end = "")
          queue.append(neighbor)                         #재귀함수의 시작점. s노드(nodes[2])를 시작으로 이전의 행위를 반복
    print("")

  def dfs(self):
    if len(self.nodes) > 0:                                 # 반복작업을 위한 노드세팅 초기화.(모두 깃발내려)
      for node in self.nodes:
        node.set_visited(False)
        
    stack = []
    self.nodes[0].set_visited(True)                         #첫번째노드(A)를 지정. (A노드 깃발올려
    print(self.nodes[0].get_data(), end="")
    stack.append(self.nodes[0])
    while len(stack) > 0:
      for neighbor in stack[len(stack)-1].get_neighbors():  #일단 현재 지정노드의 이웃들을 파고든다. 깊이우선탐색의 특징
        if not neighbor.get_visited():                      #더이상 이웃들이 없는경우(1. B 2. D) 너네들은 깃발올려!(더이상 안온다.)
          neighbor.set_visited(True)
          print(" - " + neighbor.get_data(), end = "")      #깃발올리고, 출력 이런작업 반복
          stack.append(neighbor)
          break
        else:
          stack.pop(len(stack)-1)                           #방문할 노드가 있으면 pop
    print("")

graph = Graph()

node_A = Node('A')
graph.add_node(node_A)
node_B = Node('B')
graph.add_node(node_B)
node_C = Node('C')
graph.add_node(node_C)
node_D = Node('D')
graph.add_node(node_D)
node_E = Node('E')
graph.add_node(node_E)
node_F = Node('F')
graph.add_node(node_F)
node_G = Node('G')
graph.add_node(node_G)
node_H = Node('H')
graph.add_node(node_H)
node_S = Node('S')
graph.add_node(node_S)

node_A.add_neighbor(node_B)
node_A.add_neighbor(node_S)

node_B.add_neighbor(node_A)                 # 같은 정보를 두번한 이유 : 그래프는 어떤 노드에서도 시작할 수 있다.

node_C.add_neighbor(node_D)
node_C.add_neighbor(node_E)
node_C.add_neighbor(node_F)
node_C.add_neighbor(node_G)
node_C.add_neighbor(node_S)

node_D.add_neighbor(node_C)

node_E.add_neighbor(node_C)
node_E.add_neighbor(node_H)

node_F.add_neighbor(node_C)
node_F.add_neighbor(node_G)

node_G.add_neighbor(node_F)
node_G.add_neighbor(node_S)

node_H.add_neighbor(node_E)
node_H.add_neighbor(node_G)

node_S.add_neighbor(node_A)
node_S.add_neighbor(node_C)
node_S.add_neighbor(node_G)

print("[BFS result]")
graph.bfs()
print("[DFS result]")
graph.dfs()
```
