```
class Node:
  def __init__(self, data):
    self.data = data
    self.neighbors = []
    self.visited = False
    
  def add_neighbor(self, neighbor):
    self.neighbors.append(neighbor)

  def set_visited(self, visited):
    self.visited = visited

  def get_data(self):
    return self.data

  def get_neighbors(self):
    return self.neighbors

  def get_visited(self):
    return self.visited


class Graph:
  def __init__(self):
    self.nodes = []             #노드 여러개를 위한 리스트
    self.result = []            # just output 결과를 출력하기위한 리스트
    self.stack = []             # log to how gone there
    self.loging = True          # or False

  def topological_sort(self):
    self.half_line_flag()
    self.recursive()
    self.finished()

    
  def add_node(self, node):
    self.nodes.append(node)

  def half_line_flag(self):
    if len(self.nodes) > 0:                    #노드가 있을경우 
      for node in self.nodes:
        node.set_visited(False)                #전부 초기화(깃발내려!)
        
#   def debug(self):
#     rs = ""
#     for x in self.stack:
#       rs += x.data
#     return rs
        
  def activation_function(self, node):                # activat 갈곳이 없으면 result list에 append
    choice = node.get_neighbors()
    node.set_visited(True)                            #무조건적인 흔적 남김
    is_have = 0;
    for i in choice:                                  #방향확인여부
      if(False == i.get_visited()):
        is_have +=1                                   #만약 내가 가지않은 곳이 있다면(Flase일경우) +1 = 내가 앞으로 가야할 곳
      
#     if self.loging:
#       print("node name : ", node.data, "stack", self.debug(), "forward", "is_have", is_have)  
    if is_have == 0:                                  # 값초기화
#       print("output push", node.data)               # 갈곳이 없으면 push => result list에 push
      self.result.append(node.data)
      self.back_function()                            #back cycle (재귀함수)
    else:
      self.stack.append(node)
      for i in choice:                                #choice에서 false인 노드를 찾아간다.()
        if(False == i.get_visited()): 
          self.activation_function(i)
      
  def back_function(self):
    if(len(self.stack) == 0):
      return
    
    node = self.stack.pop()
#     if self.loging:
# #       print("node name : ", node.data, "stack", self.debug(), "backward")
    self.activation_function(node)
      
        
  def recursive(self):
    if len(self.nodes) > 0:
      
        for node in self.nodes:
          if node.get_visited() == False:      
            self.activation_function(node)
  
  def finished(self):                                  #result 값 출력 함수
    rs = ""
    for index in range(len(self.result), 0, -1):
      rs += self.result[index-1]
    print(rs)     
    
    
graph = Graph()


node_A = Node('A')
graph.add_node(node_A)
node_B = Node('B')
graph.add_node(node_B)
node_C = Node('C')
graph.add_node(node_C)
node_D = Node('D')
graph.add_node(node_D)
node_E = Node('E')
graph.add_node(node_E)
node_F = Node('F')
graph.add_node(node_F)
node_G = Node('G')
graph.add_node(node_G)
node_H = Node('H')
graph.add_node(node_H)
node_I = Node('I')
graph.add_node(node_I)
node_J = Node('J')
graph.add_node(node_J)

node_A.add_neighbor(node_B)
node_A.add_neighbor(node_F)

node_B.add_neighbor(node_H)

node_D.add_neighbor(node_C)
node_D.add_neighbor(node_E)
node_D.add_neighbor(node_I)

node_E.add_neighbor(node_I)

node_G.add_neighbor(node_A)
node_G.add_neighbor(node_B)
node_G.add_neighbor(node_C)

node_I.add_neighbor(node_C)

node_J.add_neighbor(node_E)



graph.topological_sort()
```
