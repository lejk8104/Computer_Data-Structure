```
# Stack 교수님 버전

class Node:
  
  def __init__(self, data):
    self.data = data
    self.next = None
    
  def get_data(self):
    return self.data
  
  def get_next(self):
    return self.next
  
  def set_next(self, next):
    self.next = next
    
class Stack:
  
  def __init__(self):
    self.top = None
    self.size = 0
  
  def push(self, data):
    new_node = Node(data)
    self.size += 1
    if self.top is None:
      self.top = new_node
    else:
      new_node.set_next(self.top)
      self.top = new_node

  def print_stack(self):
    print(">> current stack")
    node = self.top
    while node is not None:
      print(node.get_data())
      node = node.get_next()
    print(">>>>>>>>>>>>>>>>")
  
  def pop(self):
    if self.size == 0:
      node = None
    
    node = self.top
    if node is not None:
      self.top = node.get_next()
      self.size -= 1
    return node
  
  def get_size(self):
    return self.size

  
stack = Stack()
stack.push("Apple")
stack.push("Kiwi")
stack.push("Banana")
stack.print_stack()

node = stack.pop()
print("Dequeue: " + node.get_data())
stack.print_stack()
node = stack.pop()
print("Dequeue: " + node.get_data())
stack.print_stack()

stack.push("Melon")
stack.print_stack()

node = stack.pop()
print("Dequeue: " + node.get_data())
stack.print_stack()

node = stack.pop()
print("Dequeue: " + node.get_data())
stack.print_stack()

node = stack.pop()
if node is None:
  print("Empty Here")
else:
  print("Dequeue: " + node.get_data())

node = stack.pop()
if node is None:
  print("Empty Here")
else:
  print("Dequeue: " + node.get_data())
stack.push("Melon")
stack.print_stack()
```

```
# rear와 front 는 포인터(가르키는 것)
# 노드간의 링크의 방향성과 다름
# dequeue
# front가 사라지면서 이전꺼를 가르키면서 사라짐

#rear가 새로 가르키는 얘를 새롭게 들어온 멜론 노드를 가르키고, rear가 앞으로 나만 가르키면 됨?
class Node:
  
  def __init__(self, data):
    self.data = data
    self.next = None
    
  def get_data(self):                     #문법상 다음것을 지정하는것이 맞다고 판단하여 코드를 수정했습니다.
    return self.data
    
  
  def get_next(self):
    return self.next
  
  def set_next(self, next):
    self.next = next

  
class Queue:
  def __init__(self):
    self.front = None
    self.rear = None
    self.size = 0
  
  def enqueue(self, data):
    new_node = Node(data)
    self.size += 1
    if self.front is None:                 #만약 Front 데이터의 값이 비어있다면 => 최초 입력데이터(new node)
      self.front = new_node
#      print(new_node.data, "f")           Test data F
    else:
        self.rear.set_next(new_node)       #만약 데이터가 이미 존재한다면. 다음 Rear데이터를 지정하도록
    self.rear = new_node
#    print(self.rear.data, "r")            Test data R
    
  def print_queue(self):
    print(">> current list")
    node = self.front
#    print(self.front.data ,"Test print")  Test data front가 제대로 출력되는지 확인
    while node is not None:
      print(node.get_data())
      node = node.get_next()
    print(">>>>>>>>>>>>>>>>")
  
  def dequeue(self):
    if self.size == 0:
      node = None
    node = self.front
#     print(node.data, 1)                   Test data 현재값
#     print(node.next.data, 2)              Test data 다음을 지칭하는 값
#      print(self.size, "size?")            Test Size가 제대로 감소되는지?
    if node is not None:
      self.front = node.get_next()          #데이터가 Dequeue 된다면 그다음 데이터를 지정하도록
      self.size -= 1
    else:                                    #만약 노드가 비어있다면 (size =0)
      node = None                            #아무것도 출력하지마세요.
    return node    
  
  
queue = Queue()
queue.enqueue("Apple")
queue.enqueue("Kiwi")
queue.enqueue("Banana")
queue.print_queue()

node = queue.dequeue()
if node is None:
  print("Empty Here")
else:
  print("Dequeue: " + node.get_data())
node = queue.dequeue()
if node is None:
  print("Empty Here")
else:
  print("Dequeue: " + node.get_data())
queue.enqueue("Melon")
queue.print_queue()
node = queue.dequeue()
if node is None:
  print("Empty Here")
else:
  print("Dequeue: " + node.get_data())
node = queue.dequeue()
if node is None:
  print("Empty Here")
else:
  print("Dequeue: " + node.get_data())
node = queue.dequeue()
if node is None:
  print("Empty Here")
else:
  print("Dequeue: " + node.get_data())
  
```
